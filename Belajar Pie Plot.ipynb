```python
import matplotlib
import matplotlib.pyplot as plt

print(matplotlib.__version__)
```

```python
#Simple Pie Plot

kategori= ['A','B','C','D','E']
data = [215,130,245,210,225]
```

```python
plt.pie(data, #berasosiasi dengan integer 
        labels=kategori, #berasosiasi dengan kategori (opsional bisa dipakai bisa tidak)
        autopct='%1.1f%%', #digunakan untuk menampilkan presentase dari setiap area pada Pie Plot ini (opsional).
        #kalau tidak dipakai tidak terlihat angkanya
        startangle=90) #digunakan untuk melakukan rotasi/sudut pada Pie Plot.

plt.title('Contoh Pie Plot') #untuk menampilkan judul pada Pie Plot

plt.show()
```

```python
#Pengaturan Legend dan Color

kategori= ['A','B','C','D','E']
data = [215,130,245,210,225]
warna = ['pink','cyan','#e03364','yellowgreen','skyblue'] #merepresentasikan warna dari setiap isi parameter kategori
```

```python
plt.pie(data, #berasosiasi dengan nilai integer.
        labels=kategori, #berasosiasi dengan parameter kategori dan sifatnya opsional.
        colors=warna, #berasosiasi dengan parameter warna.
        autopct='%1.1f%%', #digunakan untuk menampilkan presentase dari setiap area pada Pie Plot ini dan sifatnya juga opsional.
        startangle=90) #digunakan untuk melakukan rotasi?sudut pada Pie Plot.

plt.legend() #untuk menampilkan keterangan warna Pie Plot
plt.title('Contoh Pie Plot') 

plt.show()
```

```python
#Exploded Pie Plot

kategori= ['A','B','C','D','E']
data = [215,130,245,210,225]
warna = ['pink','cyan','#e03364','yellowgreen','skyblue'] #merepresentasikan warna dari setiap isi parameter kategori
explode_var =[0.,0.,0.2,0.,0.8] #menampung nilai ledakan yang berkorelasi dengan kategori.
```

```python
plt.pie(data, #berasosiasi dengan sekumpulan nilai integer.
        labels=kategori, #berasosiasi dengan parameter kategori dan sifatnya opsional.
        colors=warna, #berasosiasi dengan parameter warna.
        explode = explode_var, #berkolerasi dengan parameter explode_var.
        autopct='%1.1f%%', 
        startangle=90) #digunakan untuk melakukan rotasi/sudut pada Pie Plot.

plt.title('Contoh Exploded Pie Plot') 

plt.show()
```
